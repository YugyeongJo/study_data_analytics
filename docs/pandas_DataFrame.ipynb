#### row(index) 단위 : iloc vs loc
- iloc : 위치값으로 접근
- loc : 명칭으로 접근

```python
df_saledays.index = ['a','b','c','d']
```

```python
df_saledays
```

```python
# df_saledays.iloc[2]
df_saledays.loc['c']
```

#### index와 column을 같이 추출할때 사용

```python
df_saledays.loc['c', '판매일자'] # indexes, columns
```

```python
df_saledays.loc[:, '판매일자'] # : = 전체를 의미
```

```python
df_saledays.loc[:, '판매일자':] # column에서 판매일자 뒤 모든 값
```

### 정렬

```python
df_saledays.sort_values(by='가격', ascending=True)  # 단일 변수일 때
```

```python
df_saledays.sort_values(by=['가격', '반품여부'], ascending=[False, True])  # 다 변수일 때 # by의 숫자와 ascendinf의 숫자를 맞춰줘야 작동함
```
