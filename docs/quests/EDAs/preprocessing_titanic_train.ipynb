## 1. Age' 열에서 결측치를 평균 나이로 채우기.

```python
df_TFD_01 = df_TFD.copy()
```

```python
df_TFD_01.isnull().sum()
```

```python
mean_Age = df_TFD_01.describe().loc['mean', 'Age']
```

```python
df_TFD_01['Age'].fillna(mean_Age, inplace=True)
```

```python
df_TFD_01.isnull().sum()
```

## 4. 'Fare' 열에서 평균보다 3 표준편차(mean + 3 * std) 이상 높은 이상치를 가진 행들을 찾기.

```python
df_TFD_04 = df_TFD.copy()
```

```python
df_TFD_04_Fare_mean = df_TFD_04.describe().loc["mean", 'Fare']
df_TFD_04_Fare_std = df_TFD_04.describe().loc["std", 'Fare']
df_TFD_04_Fare_standard_deviation = df_TFD_04_Fare_mean + (3*df_TFD_04_Fare_std)

df_TFD_04[df_TFD_04['Fare'] >= df_TFD_04_Fare_standard_deviation]
```
