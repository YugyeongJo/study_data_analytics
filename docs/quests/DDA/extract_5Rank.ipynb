```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df_ShoppingMall = pd.read_csv('/content/ShoppingMallDeliveryWithDate.csv')
df_ShoppingMall[100:104]
```

```python
df_ShoppingMall['결제수단'].value_counts()
```

```python
sns.countplot(data=df_ShoppingMall, y='결제수단')
```

```python
series_payment = df_ShoppingMall['결제수단'].value_counts()
payment_list_high  = series_payment.iloc[:3].index.to_list()
sns.countplot(data=df_ShoppingMall, y='결제수단', order=payment_list_high)
```

```python
payment_list_low = series_payment.iloc[-3:].index.to_list()
sns.countplot(data=df_ShoppingMall, y='결제수단', order=payment_list_low)
```

```python
df_ShoppingMall['주문경로'].value_counts()
```

```python
series_orderway = df_ShoppingMall['주문경로'].value_counts()
orderway_list_high  = series_orderway.iloc[:1].index.to_list()
sns.countplot(data=df_ShoppingMall, y='주문경로', order=orderway_list_high)
orderway_list_low = series_orderway.iloc[-1:].index.to_list()
sns.countplot(data=df_ShoppingMall, y='주문경로', order=orderway_list_low)
```
